```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
```

```python
df1 = pd.read_csv("train_E6oV3lV.csv")[['tweet', 'label']]
df1.rename(columns={'label': 'class'}, inplace=True)

df2 = pd.read_csv("train.csv")[['tweet', 'class']]
```

```python
# Step 4: Combine both datasets
combined_df = pd.concat([df1, df2], ignore_index=True)
```

```python
# Step 5: Apply random oversampling to balance all classes
X = combined_df['tweet']
y = combined_df['class']
```

```python
oversampler = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = oversampler.fit_resample(X.values.reshape(-1, 1), y)
```

```python
# Step 6: Create a new balanced DataFrame
balanced_df = pd.DataFrame({'tweet': X_resampled.flatten(), 'class': y_resampled})
```

```python
balanced_df.to_csv("balanced_dataset1.csv", index=False)
```
